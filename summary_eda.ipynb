```python
from matplotlib import pyplot as plt
from src.data_loader import load_df
```

```python
df = load_df()
df.columns
```

```python
df["summ_sentence_len"] = df["highlights"].apply(lambda x: len(x.split(".")))
df["art_sentence_len"] = df["article"].apply(lambda x: len(x.split(".")))
df["summ_word_len"] = df["highlights"].apply(lambda x: len(x.split(" ")))
df["art_word_len"] = df["article"].apply(lambda x: len(x.split(" ")))
```

```python
df["summ_sentence_len"].hist(bins=10)
plt.title("Sentences Per Summary Distribution")
plt.xlabel("Number of Sentences in a Summary")
plt.ylabel("Count")

plt.show()
```

```python
df["summ_word_len"].hist(bins=10)
plt.xlabel("Number of Words in a Summary")
plt.ylabel("Count")
plt.title("Words Per Summary Distribution")
plt.show()
```

### Insights to Apply to Zero Shot Summarization Task
* Majority of summaries are 5 sentences in length. Baseline summary (first X sentences) will be set to 5.
* Majority of summaries are under 80-100 words. Most summaries are ~30-50 words.
